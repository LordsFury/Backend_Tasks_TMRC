```python

# What is Django?
# Django is a high-level Python web framework that enables developers to build secure, scalable, and maintainable web applications quickly and efficiently.
# Key Features of Django are:
# Comes with built-in tools such as authentication, an admin panel, forms, and ORM
# MVT architecture
# Secure
# Rapid development
# Scalable


# Diff b/w Django and djangorestframework?

# Django:
# A full-stack web framework for building traditional web applications.
# Focuses on rendering HTML templates, managing forms, handling HTTP requests/responses, and interacting with databases.
# Best suited for websites that return HTML (like blogs, e-commerce sites, etc.).
# Used for building server-rendered websites, Admin dashboards, and User authentication.
    
# Django REST Framework (DRF):
# An extension of Django designed to build RESTful APIs.
# Allows you to build backend APIs that return data (usually in JSON) instead of HTML.
# Makes it easy to expose Django models to other clients (like frontend apps built in React, Angular, or mobile apps).
# Used for building APIs, creating JSON-based endpoints, and supporting CRUD operations over HTTP


# Django architecture?
# Django follows the MVT architecture. It is similar to MVC (Model-View-Controller) but with slight naming differences.
    
# Model:
# Handles the data and business logic.
# Defines your database schema using Python classes.
# Interacts with the database using Django’s ORM.
    
# View:
# Controls the logic and handles requests/responses.
# Fetches data from the model and sends it to the template.
# Returns an HTTP response (HTML, JSON, etc.).
    
# Template:
# Manages the presentation layer (UI).
# Displays data to the user.
# Uses Django’s template language to embed dynamic content.

    
# Django code level flow?
# User Sends Request (Browser/Frontend):
# User accesses a URL like:
# http://localhost:8000/books/
# URL Routing — urls.py:
# The URL is matched with a route in the urls.py file.
# View Logic — views.py:
# When the URL matches, Django calls the corresponding view function or class.
# Data Access — models.py:
# The view fetches data from the database via the model.
# Template Rendering — templates/books.html:
# The view passes data (books) to the template for rendering HTML.
# HTTP Response:
# The template is rendered to an HTML page, and Django sends it back as an HTTP response to the browser.

    
# Django request response cycle?
# User Sends HTTP Request
# A user opens a browser and accesses a URL, e.g.,
# http://example.com/books/
# The browser sends an HTTP request to the Django server.
# WSGI/ASGI Gateway Receives Request
# Django uses a WSGI/ASGI server (like Gunicorn, Daphne, or built-in dev server) to handle the incoming request.
# Django Creates HttpRequest Object
# Django creates an HttpRequest object from the raw request.
# This object contains metadata like headers, method, user info, GET/POST data, etc.
# Middleware Processing
# The request passes through a middleware stack (defined in settings.py).
# Middleware can  modify the request or response, handle authentication, sessions, security, etc. short-circuit and return a response directly
# URL Dispatcher (Routing)
# Django uses urls.py to match the request path to a view function.
# View Function Execution
# The matched view is called with the HttpRequest object.
# It processes data, often interacts with the Model (database), and returns an HttpResponse.
# Browser Request
#      ↓
# WSGI/ASGI Server
#      ↓
# Django Middleware (Request Phase)
#      ↓
# URL Resolver → Matched View
#      ↓
# View executes → Model (optional) → Template (optional)
#      ↓
# HttpResponse Created
#      ↓
# Django Middleware (Response Phase)
#      ↓
# WSGI/ASGI Server → Back to Browser


# Diff between django app and django project?
# Django Project:
# The entire web application
# Manages the configuration and settings
# django-admin startproject myproject

# Django App:
# A modular component inside the project
# Handles specific features or functionality
# python manage.py startapp myapp


# Django app and project file structure?
# Django Project:
# django-admin startproject myproject
# myproject/
# │
# ├── manage.py                
# ├── myproject/               
# │   ├── __init__.py
# │   ├── settings.py          
# │   ├── urls.py           
# │   ├── asgi.py              
# │   └── wsgi.py             

# Django App:
# python manage.py startapp blog
# myproject/
# │
# ├── manage.py
# ├── myproject/
# │   ├── __init__.py
# │   ├── settings.py
# │   ├── urls.py
# │   ├── asgi.py
# │   └── wsgi.py
# │
# ├── blog/                     
# │   ├── __init__.py
# │   ├── admin.py              
# │   ├── apps.py              
# │   ├── models.py            
# │   ├── tests.py             
# │   ├── views.py              
# │   ├── urls.py          
# │   └── migrations/        
# │       └── __init__.py

    
# Django admin panel?
# The Django admin panel is a powerful, built-in web interface that allows you to manage your application’s data using a user-friendly dashboard, eliminating the need to write custom HTML or forms.
# It's automatically available once you create a Django project.
# Lets you create, read, update, and delete (CRUD) records for your models.
# Great for managing users, blog posts, products, etc.
# Only accessible by superusers (admins).
```
